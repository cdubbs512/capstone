# Define and use defined csv_cleaner function to clean dataframes

```python

def csv_cleaner(dataframe):

    try:
        dataframe['goals'] = dataframe['goals'].apply(lambda x: str(x).replace('-', str(0)))
        dataframe['goals'] = dataframe['goals'].astype(float)


        dataframe['assists'] = dataframe['assists'].apply(lambda x: str(x).replace('-', str(0)))
        dataframe['assists'] = dataframe['assists'].astype(float)


        dataframe['yellow_card'] = dataframe['yellow_card'].apply(lambda x: str(x).replace('-', str(0)))
        dataframe['yellow_card'] = dataframe['yellow_card'].astype(float)


        dataframe['second_yellow'] = dataframe['second_yellow'].apply(lambda x: str(x).replace('-', str(0)))
        dataframe['second_yellow'] = dataframe['second_yellow'].astype(float)


        dataframe['red_cards'] = dataframe['red_cards'].apply(lambda x: str(x).replace('-', str(0)))
        dataframe['red_cards'] = dataframe['red_cards'].astype(float)


        dataframe['substituted_on'] = dataframe['substituted_on'].apply(lambda x: str(x).replace('-', str(0)))
        dataframe['substituted_on'] = dataframe['substituted_on'].astype(float)


        dataframe['substituted_off'] = dataframe['substituted_off'].apply(lambda x: str(x).replace('-', str(0)))
        dataframe['substituted_off'] = dataframe['substituted_off'].astype(float)


        dataframe['ppm'] = dataframe['ppm'].apply(lambda x: str(x).replace(',', '.'))
        dataframe['ppm'] = dataframe['ppm'].astype(float)


        dataframe['appearances'] = dataframe['appearances'].apply(lambda x: str(x).replace('Was not used during this season', str(0)))
        dataframe['appearances'] = dataframe['appearances'].apply(lambda x: str(x).replace('Not in squad during this season', str(0)))
        dataframe['appearances'] = dataframe['appearances'].astype(float)


        dataframe['in_squad'] = dataframe['in_squad'].apply(lambda x: str(x).replace('-', str(0)))
        dataframe['in_squad'] = dataframe['in_squad'].apply(lambda x: str(x).replace('Was not used during this season', str(0)))
        dataframe['in_squad'] = dataframe['in_squad'].apply(lambda x: str(x).replace('Not in squad during this season', str(0)))
        dataframe['in_squad'] = dataframe['in_squad'].astype(float)


        dataframe['number'] = dataframe['number'].apply(lambda x: str(x).replace('-', str(-999)))
        dataframe['number'] = dataframe['number'].astype(int)

        
        dataframe['age'] = dataframe['age'].apply(lambda x: str(x).replace('-', str(-999)))
        dataframe['age'] = dataframe['age'].apply(lambda x: str(x).replace('†', ''))
        dataframe['age'] = dataframe['age'].astype(float)
       
    
        dataframe['height_m'] = dataframe['height_m'].apply(lambda x: str(x).replace(',', '.').replace('m', '').strip())
        dataframe['height_m'] = dataframe['height_m'].apply(lambda x: str(x).replace(' ', ''))
        #dataframe['height_m'] = dataframe['height_m'].apply(lambda x: str(x).replace('', str(0)))
        dataframe['height_m'] = dataframe['height_m'].apply(lambda x: str(x).replace('', ''))
        dataframe['height_m'] = dataframe['height_m'].convert_objects(convert_numeric=True)
        dataframe['height_m'] = dataframe['height_m'].astype(float)
    
        
        dataframe['minutes_played'] = dataframe['minutes_played'].apply(lambda x: str(x).replace('.', '').replace("'", "").strip())
        dataframe['minutes_played'] = dataframe['minutes_played'].apply(lambda x: str(x).replace('-', str(0)))
        dataframe['minutes_played'] = dataframe['minutes_played'].astype(float)


        dataframe['transfer_fee'] = dataframe['transfer_fee'].apply(lambda x: str(x).replace('.', ''))
        dataframe['transfer_fee'] = dataframe['transfer_fee'].apply(lambda x: str(x).replace('£', '').replace("'", ""))
        dataframe['transfer_fee'] = dataframe['transfer_fee'].apply(lambda x: str(x).replace('m', '0000'))
        dataframe['transfer_fee'] = dataframe['transfer_fee'].apply(lambda x: str(x).replace('k', '000'))
        dataframe['transfer_fee'] = dataframe['transfer_fee'].apply(lambda x: str(x).replace('-', str(0)))
        dataframe['transfer_fee'] = dataframe['transfer_fee'].apply(lambda x: str(x).replace('?', str(0)))
        dataframe['transfer_fee'] = dataframe['transfer_fee'].apply(lambda x: str(x).replace('Free', str(0)))
        dataframe['transfer_fee'] = dataframe['transfer_fee'].apply(lambda x: str(x).replace('nan', str(0)))
        dataframe['transfer_fee'] = dataframe['transfer_fee'].astype(float)


        dataframe['market_value'] = dataframe['market_value'].apply(lambda x: str(x).replace('.', ''))
        dataframe['market_value'] = dataframe['market_value'].apply(lambda x: str(x).replace('£', '').replace("'", ""))
        dataframe['market_value'] = dataframe['market_value'].apply(lambda x: str(x).replace('m', '0000'))
        dataframe['market_value'] = dataframe['market_value'].apply(lambda x: str(x).replace('k', '000'))
        dataframe['market_value'] = dataframe['market_value'].apply(lambda x: str(x).replace('?', str(0)))
        dataframe['market_value'] = dataframe['market_value'].apply(lambda x: str(x).replace('-', str(0)))
        dataframe['market_value'] = dataframe['market_value'].apply(lambda x: str(x).replace('nan', str(0)))
        dataframe['market_value'] = dataframe['market_value'].astype(float)


        dataframe.to_csv('dataframe_cleaned.csv')
        print(dataframe.shape)

        return dataframe
        
    except ValueError:
        print("Oops!  There may be an error.")

```
